# Why Python? 

- **Easy to learn**
    - Simple syntax and vocabulary, without a lot of clutter
    - Most things have "One obvious way to do it"

- **Encourages logical thinking**
    - Block structure, with clear visual hierarchy
    - Lots of built-in data types, flexible 
    - "Coding at the speed of thought"

- **Amazing industry support**
    - Lots and lots of free/open source software
    - Easy to "glue together" whatever you need

It is enough to simply complete them the best you can. **Quizzes and the final project, on the other hand, will each receive a letter grade that factors directly in your course average.**   

**For more details see the course syllabus.**

# Grading Scale
**Every graded assignment will be scored and then normalized** using the following formula:  
> QP = 3.5 + ½ (x-μ)/σ,   

where
- x is the student’s raw score for the assignment
- μ and σ are the class average and standard deviation for the assignment
- QP is the quality points (~ GPA) used in the course average

Letter grades are then 3.67+ → A, 3.34-3.66 → A-, …
